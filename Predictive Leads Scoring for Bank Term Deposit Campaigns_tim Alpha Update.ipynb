# **5. Evaluation Metrics**

Evaluasi metrik merupakan proses untuk menilai seberapa baik model prediksi bekerja dalam membedakan antara nasabah yang berpotensi menyetujui deposito dan yang tidak. Metrik digunakan sebagai acuan objektif untuk mengukur kualitas model sehingga keputusan bisnis dapat dibuat secara lebih terukur. Dalam konteks klasifikasi biner seperti kasus kampanye deposito berjangka, evaluasi performa model sangat bergantung pada struktur dasar yang disebut Confusion Matrix.

Confusion Matrix membagi hasil prediksi model menjadi empat komponen utama:  
- True Positive (TP) — model memprediksi "setuju" dan benar-benar "setuju".  
- True Negative (TN) — model memprediksi "tidak setuju" dan benar-benar "tidak setuju".  
- False Positive (FP) — model memprediksi "setuju" padahal sebenarnya "tidak setuju".  
- False Negative (FN) — model memprediksi "tidak setuju" padahal sebenarnya "setuju".

Berbagai metrik evaluasi diturunkan dari empat komponen tersebut:

### **Precision**
Formula (plain text):  
Precision = TP / (TP + FP)

### **Recall**
Formula (plain text):  
Recall = TP / (TP + FN)

### **F-Score (general formula)**
Formula (plain text):  
F_beta = (1 + beta^2) * (Precision * Recall) / ((beta^2 * Precision) + Recall)

---

## **False Positive (FP) dan False Negative (FN) dalam Konteks Kampanye Deposito**

### False Positive (FP)  
Model memprediksi nasabah akan setuju, tetapi sebenarnya tidak tertarik.  
Dampak:  
- Biaya operasional meningkat karena agen menghubungi leads yang tidak produktif.  
- Waktu agen terbuang.  
- Cost per Acquisition meningkat.

### False Negative (FN)  
Model memprediksi nasabah tidak akan setuju, tetapi sebenarnya berminat.  
Dampak:  
- Potensi pendapatan deposito hilang.  
- Penurunan tingkat konversi keseluruhan.  
- Leads bernilai tinggi tidak dihubungi.

### Prioritas Evaluasi Model

SSebagai tim Data Scientist, kami memprioritaskan pengurangan **False Negative (FN)** untuk memastikan bahwa tidak ada calon nasabah potensial yang terlewat oleh sistem prediksi. False Negative terjadi ketika model memprediksi seorang nasabah tidak akan menaruh uang di deposito, padahal sebenarnya mereka berminat. Kesalahan ini berdampak langsung pada hilangnya peluang pendapatan dan menurunkan tingkat konversi kampanye.

Untuk itu, metrik utama yang menjadi fokus adalah **Recall**, yaitu kemampuan model dalam menangkap seluruh nasabah yang benar-benar berminat. Recall yang tinggi menunjukkan bahwa model berhasil meminimalkan FN, sehingga lebih banyak calon nasabah bernilai tinggi dapat dihubungi oleh agen telemarketing. Pendekatan ini memastikan bahwa kampanye berjalan lebih efektif, peluang penjualan meningkat, dan potensi pendapatan tidak terbuang.

---

## **Pemilihan Metrik Utama: Recall**

Dalam konteks kampanye telemarketing deposito berjangka, setiap **False Negative (FN)** — **yaitu nasabah yang sebenarnya berminat berlangganan tetapi diprediksi tidak berminat oleh model** — menyebabkan hilangnya peluang konversi dan potensi pendapatan. FN membuat agen melewatkan nasabah bernilai tinggi yang seharusnya dihubungi, sehingga efektivitas kampanye menurun. Oleh karena itu, sebagai tim Data Scientist, kami memprioritaskan pengurangan FN sebagai bagian dari strategi optimasi model.

Untuk mencapai tujuan tersebut, metrik utama yang digunakan adalah Recall, yaitu proporsi nasabah yang benar-benar berminat dan berhasil ditangkap oleh model.

Formula (plain text):
Recall = True Positive / (True Positive + False Negative)

Dengan menjadikan Recall sebagai metrik utama, model dioptimalkan untuk meminimalkan jumlah FN, sehingga lebih sedikit nasabah potensial yang terlewat. Pendekatan ini memastikan bahwa daftar calon nasabah yang dihasilkan lebih komprehensif, meningkatkan peluang konversi, dan memaksimalkan dampak kampanye telemarketing secara keseluruhan.

## 1.5. Membuat generational marketing (www.bai.org on “banking attitudes, generation-by-generation")

```python
REFERENCE_YEAR = 2014

def map_generation(age, reference_year=REFERENCE_YEAR):
    try:
        birth = reference_year - int(age)
    except:
        return 'unknown'

    if 1946 <= birth <= 1964:
        return 'Baby Boomer'
    elif 1965 <= birth <= 1979:
        return 'Generation X'
    elif 1980 <= birth <= 1994:
        return 'Millennials'
    elif 1995 <= birth <= 2012:
        return 'Generation Z'
    else:
        return 'Other'

if 'generation' not in df.columns and 'age' in df.columns:
    df['generation'] = df['age'].apply(map_generation)
```

Laporan BAI (penyedia layanan & advisory finansial terkait perbankan), memberikan insight perilaku bank khusus untuk:
tahun lahir = 2014-umur

1. Baby Boomers : 1946–1964 (usia 61–79 tahun)
2. Gen X        : 1965–1979 (usia 46–60 tahun)
3. Millennials  : 1980–1994 (usia 31–45 tahun)

BAI **tidak** memberikan analisis rinci untuk generasi:

- Gen Z (lahir 1995–2010) yang masih sangat muda atau belum menjadi target deposito jangka panjang
- Silent Generation (>80 tahun)
- Anak-anak (<18 tahun)

Karena dataset bank telemarketing Portugal berfokus pada produk term deposit (produk investasi berbasis bunga yang memerlukan kapasitas finansial), maka yang relevan dengan studi perilaku finansial menurut BAI.org adalah ketiga generasi tersebut.

## 3.1. Mendefinisikan model

```python
# ===============================
# Imports & Setup
# ===============================
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import make_scorer, fbeta_score, recall_score, precision_score, accuracy_score

from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, VotingClassifier
from xgboost import XGBClassifier
import lightgbm as lgb

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# ===============================
# CV & Scoring
# ===============================
RANDOM_STATE = 42

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=RANDOM_STATE
)

scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f2': make_scorer(fbeta_score, beta=2)
}

# ===============================
# Models
# ===============================
models = {
    "RandomForest": RandomForestClassifier(
        n_estimators=200,
        random_state=RANDOM_STATE,
        class_weight='balanced'
    ),

    "Bagging": BaggingClassifier(
        n_estimators=200,
        random_state=RANDOM_STATE
    ),

    "XGBoost": XGBClassifier(
        eval_metric='logloss',
        use_label_encoder=False,
        random_state=RANDOM_STATE,
        n_estimators=300,
        learning_rate=0.05
    ),

    "LightGBM": lgb.LGBMClassifier(
        random_state=RANDOM_STATE,
        n_estimators=300
    ),

    "Voting": VotingClassifier(
        estimators=[
            ('rf', RandomForestClassifier(
                n_estimators=200,
                random_state=RANDOM_STATE,
                class_weight='balanced'
            )),
            ('bag', BaggingClassifier(
                n_estimators=200,
                random_state=RANDOM_STATE
            )),
            ('xgb', XGBClassifier(
                eval_metric='logloss',
                use_label_encoder=False,
                random_state=RANDOM_STATE,
                n_estimators=300,
                learning_rate=0.05
            ))
        ],
        voting='soft'
    )
}
```

## 3.2. Cross-Validation dengan SMOTE

Evaluasi dilakukan menggunakan Stratified 5-Fold Cross-Validation dengan SMOTE untuk menangani class imbalance.
Metrik utama yang digunakan adalah F2-score, karena fokus analisis adalah memaksimalkan recall (mengurangi false negative pada calon nasabah potensial).

```python
# ===============================
# Cross-Validation with SMOTE
# ===============================
results = []

for name, model in models.items():
    print(f"\nRunning CV for: {name}")

    pipe = ImbPipeline(steps=[
        ('preprocess', preprocess),   # ColumnTransformer
        ('smote', SMOTE(random_state=RANDOM_STATE)),
        ('model', model)
    ])

    cv_result = cross_validate(
        pipe,
        X_train,
        y_train,
        cv=cv,
        scoring=scoring,
        n_jobs=-1,
        return_train_score=False
    )

    results.append({
        'model': name,
        'accuracy_mean': cv_result['test_accuracy'].mean(),
        'precision_mean': cv_result['test_precision'].mean(),
        'recall_mean': cv_result['test_recall'].mean(),
        'f2_mean': cv_result['test_f2'].mean()
    })

results_df = pd.DataFrame(results).sort_values(
    by='f2_mean', ascending=False
)

results_df
```

Berdasarkan F2-score, model RandomForest menunjukkan performa terbaik:

Memiliki recall tertinggi (36.2%)

F2-score paling tinggi → cocok untuk use case lead scoring / campaign targeting

Lebih seimbang antara precision dan recall dibanding model lain

Hyperparameter tuning untuk **Random Forest** dilakukan karena hasil cross-validation awal menunjukkan bahwa model ini memiliki nilai F2 score dan Recall tertinggi, sehingga paling potensial dalam menekan False Negative (FN).
```python
from sklearn.metrics import make_scorer, recall_score
from sklearn.model_selection import RandomizedSearchCV
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier

# Recall scorer (minimize FN)
recall_scorer = make_scorer(recall_score, pos_label=1)

rf_params = {
    'rf__n_estimators': [200, 300, 500, 800, 1000],
    'rf__max_depth': [5, 10, 20, 30, None],
    'rf__min_samples_split': [2, 5, 10],
    'rf__min_samples_leaf': [1, 2, 4],
    'rf__max_features': ['sqrt', 'log2'],
    'rf__bootstrap': [True, False]
}

pipeline_rf = ImbPipeline(steps=[
    ('preprocessor', preprocess),
    ('smote', SMOTE(random_state=42)),
    ('rf', RandomForestClassifier(random_state=42))
])

rf_search = RandomizedSearchCV(
    estimator=pipeline_rf,
    param_distributions=rf_params,
    n_iter=30,
    scoring=recall_scorer,
    cv=5,
    n_jobs=-1,
    verbose=2,
    random_state=42
)

rf_search.fit(X_train, y_train)

best_model_rf = rf_search.best_estimator_

print("Best Recall Score:", rf_search.best_score_)
print("Best Parameters:", rf_search.best_params_)
```

# 4. Cost-Benefit Calculator

Menjalankan Cost–Benefit Analysis

```python
def cost_benefit_analysis(
    y_true, 
    y_pred, 
    benefit_TP=0, 
    cost_FN=0,      # fokus utama
    cost_FP=0, 
    benefit_TN=0
):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    TN = np.sum((y_true == 0) & (y_pred == 0))

    Cost_FN_Total = FN * cost_FN
    Cost_FP_Total = FP * cost_FP

    Benefit_TP_Total = TP * benefit_TP
    Benefit_TN_Total = TN * benefit_TN

    Net_Impact = (
        Benefit_TP_Total + Benefit_TN_Total
        - (Cost_FN_Total + Cost_FP_Total)
    )

    return {
        "TP": int(TP),
        "FP": int(FP),
        "FN": int(FN),
        "TN": int(TN),
        "Cost_FN_Total": int(Cost_FN_Total),
        "Cost_FP_Total": int(Cost_FP_Total),
        "Benefit_TP_Total": int(Benefit_TP_Total),
        "Benefit_TN_Total": int(Benefit_TN_Total),
        "Net_Impact": int(Net_Impact)
    }
```

```python
y_pred = best_model_rf.predict(X_test)

results = cost_benefit_analysis(
    y_true=y_test,
    y_pred=y_pred,
    
    # Fokus utama → biaya FN
    cost_FN=30,         # biaya untuk setiap False Negative
    
    # Komponen lain (opsional)
    cost_FP=10,         # biaya FP (sekunder)
    benefit_TP=50,      # benefit jika prediksi positif benar
    benefit_TN=0        # TN dianggap netral
)

results
```

Net Impact = (Benefit TP + Benefit TN) − (Cost FN + Cost FP)

Perhitungan: 29,750 + 0 − (9,990 + 9,880) = 9,880

Interpretasi: Setelah memperhitungkan biaya FN dan FP, model masih memberikan dampak positif bersih sebesar **EUR 9,880**.

Setelah melakukan perbandingan model dan tuning yang berfokus pada minimasi kerugian False Negative (FN)—yaitu potensi profit yang hilang—model Random Forest Classifier terbukti memberikan performa terbaik dalam menyeimbangkan penangkapan nasabah potensial dan efisiensi.

```python
def predict_term_deposit(data: pd.DataFrame, model):
    """
    Return:
    - pred_label (0 / 1)
    - pred_prob (float)
    """
    pred_label = model.predict(data)[0]
    pred_prob = model.predict_proba(data)[0, 1]
    return pred_label, pred_prob
```
